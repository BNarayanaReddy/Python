##### Using loc function

```python
df.head(2)
```

```python
 df.loc[0,0] = 100   # row,column
df
```

```python
df = df.drop(0)
df = df.drop(0,axis = 1)
df
```

```python
df.loc[1][4] = 10
df.head(2)
```

```python
df.loc[[1,2],[4,3]]  #rows,columns
```

```python
df.head()
```

```python
# Manipulating the columns
df.loc[df[1] < 0.3]
```

```python
df.loc[df[1] < 0.4,]
```

```python
# Accessing using .iloc which assumes the column in index manner(0 to n-1)
df.head(5)
```

```python
df.iloc[1,1] # same as df.loc[2,2]
```
